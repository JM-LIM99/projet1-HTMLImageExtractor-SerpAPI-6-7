```python
#Ce code pour extrait la page html et l'image./this projet for exact the text and image of article/html , 이미지 추출 
#classfication des html et images/classify text and image of article./html 과 이미지 분류

pip install google-search-results # utilisation de google /use googgle research prgram/구글 리서치 API 사용

from serpapi import GoogleSearch #
search = GoogleSearch({
    "q": "mains coupées/Congo", # 
    "location": "Paris,France", # les mots de clés/The keyword is cited in the magazine’s section "Désinformation visuelle"/키워드
    "api_key": ""
  })

result = search.get_dict() #use serpapi

import json

with open("texte.txt", "w") as w:
    w.write(json.dumps(result)) #stock results
    
with open("texte.txt") as r:
    res = json.load(r)
    print(res) # visualization
    
for result in res.get("organic_results", []):
    print("title:", result.get("title"))
    print("link:", result.get("link"))
    print("\n") #Titre and link of the article

#stock files in PC
import os
from urllib.request import urlopen

os.makedirs("exemple keyword", exist_ok=True)

cpt = 0
for i in res.get("organic_results", []):
    link= i.get("link")
    if link:
        try:
            response = urlopen(link)
            html_content = response.read()
            
            cpt += 1
            folder_path = f"exemple keyword/{cpt}"
            os.makedirs(folder_path, exist_ok=True)
            
            with open(f"{folder_path}/index.html", "wb") as f:
                f.write(html_content)

            print(f"{folder_path}/index.html")

        except Exception as e:
            print(f"{link} : {e}")

# extract image and url of article             
import requests
import urllib.request
from urllib.parse import urljoin
from bs4 import BeautifulSoup
import os
num = 2
num1 = 1
img_urls =[]
base_url = "https://example.com/"


while num < 9 :
        with open(f"mains_coupées/{num}/index.html","r", encoding="utf-8") as f:
            soup = BeautifulSoup(f, "html.parser")
            img_tags = soup.find_all("img")
            for i in img_tags:
                src = i.get("src")
                if not src:
                    continue
                try:
                    savename = f"mains_coupées/{num}/image{num1}.jpg"
                    urllib.request.urlretrieve(src, savename)
                    
                    print(f"stock :{num1}")
                    num1 += 1
                except Exception as e:
                    print("fail")
                full_url = urljoin(base_url, src)
                img_urls.append(full_url)
            print("[prochaine]")
            num += 1
            
with open("img_urls.txt", "w", encoding="utf-8") as f:
    for url in img_urls:
        f.write(url + "\n")

#프로젝트 6월 7월
            
```
